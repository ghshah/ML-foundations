```python
import os
from keras.preprocessing import image

folders = os.listdir('Train')

image_data = []
labels = []
count = 0

for ix in folders:
  path = os.path.join("Train", ix)
  print(path, count)
  for im in os.listdir(path):
    try:
      img = image.load_img(os.path.join(path,im), target_size=(224,224))
      img_array = image.img_to_array(img)
      image_data.append(img_array)
      labels.append(count)
    except:
      pass
  count += 1
```

```python
import random

combined_dataset = list(zip(image_data, labels))
random.shuffle(combined_dataset)
image_data[:], labels[:] = zip(*combined_dataset)
```

```python
import numpy as np
from keras.utils import to_categorical
X_train=np.array(image_data)
Y_train=np.array(labels)

Y_train = to_categorical(Y_train)
print(Y_train)
```

```python
# from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.applications.mobilenet import MobileNet
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import *
from tensorflow.keras.models import Model
import numpy as np

model = MobileNet(include_top= False, weights= 'imagenet', input_shape=(224,224,3))
print(model.summary())
```

```python
av1 = GlobalAveragePooling2D()(model.output)

fc1 = Dense(256, activation='relu')(av1)

d1 = Dropout(0.5)(fc1)

# fc2 = Dense(128, activation='relu')(d1)

# d2 = Dropout(0.5)(fc2)

# fc3 = Dense(64, activation='relu')(d2)

# d3 = Dropout(0.5)(fc3)

fc4 = Dense(10, activation='softmax')(d1)
```

```python
model_new = Model(inputs = model.input, outputs = fc4)
model_new.summary()
```

```python
adam = Adam(learning_rate=0.00003)
model_new.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
```

```python
for ix in range(len(model_new.layers)):
  print(ix, model_new.layers[ix])
```

```python
for ix in range(72):
  model_new.layers[ix].trainable = False

print(model_new.summary())
```

```python
hist = model_new.fit(X_train, Y_train, shuffle= True, batch_size=16, epochs=8, validation_split=0.20)
```
